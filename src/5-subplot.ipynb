### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(0, 10, 1)
y = x ** 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title("Line Plot")
axs[0, 0].set_xlabel("X-axis")
axs[0, 0].set_ylabel("Y-axis")

# Bar plot
axs[0, 1].bar(categories, values, color=['blue', 'red', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart")

# Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple', marker='o')
axs[1, 0].set_title("Scatter Plot")
for i in range(len(scatter_x)):
    axs[1, 0].annotate(f"({scatter_x[i]}, {scatter_y[i]})", (scatter_x[i], scatter_y[i]), textcoords="offset points", xytext=(0,5), ha='center')

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140)
axs[1, 1].set_title("Pie Chart")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

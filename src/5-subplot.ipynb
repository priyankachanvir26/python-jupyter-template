### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

# Define data
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Check if data is valid
if len(x) == len(y) and len(categories) == len(values):
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))

    # Line Plot
    axs[0, 0].plot(x, y, marker='o', linestyle='-', color='blue')
    axs[0, 0].set_title("Line Plot")

    # Bar Plot
    axs[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'orange'])
    axs[0, 1].set_title("Bar Chart")

    # Scatter Plot
    axs[1, 0].scatter(x, y, color='purple')
    axs[1, 0].set_title("Scatter Plot")

    # Pie Chart
    axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%')
    axs[1, 1].set_title("Pie Chart")

    plt.tight_layout()
    plt.show()
else:
    print("Error: Data is not valid for subplots.")
```

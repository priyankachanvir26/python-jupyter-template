### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

class DataMismatchError(Exception):
    """Custom exception for mismatched data lengths."""
    pass

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Error Handling
if len(x) != len(y):
    raise DataMismatchError("x and y lists must have the same length.")

# Plot
plt.figure(figsize=(7, 5))
plt.scatter(x, y, color='red', edgecolor='black', s=100)

# Annotations
for i, j in zip(x, y):
    plt.text(i, j, f'({i},{j})', fontsize=10, ha='right')

# Customization
plt.title("Scatter Plot with Annotations", fontsize=14, fontweight='bold')
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

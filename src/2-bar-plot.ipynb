### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'orange']

# Create bar plot
plt.bar(categories, values, color=colors)

# Customize the plot
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the plot
plt.show()
```

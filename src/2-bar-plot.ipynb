### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

class DataMismatchError(Exception):
    """Custom exception for mismatched data lengths."""
    pass

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['blue', 'green', 'red', 'purple']

# Error Handling
if len(categories) != len(values):
    raise DataMismatchError("Categories and values must have the same length.")

# Plot
plt.figure(figsize=(7, 5))
plt.bar(categories, values, color=colors)

# Customization
plt.title("Bar Plot of Categories", fontsize=14, fontweight='bold')
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

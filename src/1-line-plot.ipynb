### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)  # x values from 0 to 10
y = x**2  # Squaring x values

# Validate input data
if len(x) == len(y) and len(x) > 0:
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, color='b', linestyle='-', marker='o', label="y = x²")
    
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("Line Plot of y = x²")
    plt.grid(True)
    plt.legend()
    plt.show()
else:
    print("Error: x and y must have the same length and should not be empty.")
```

### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

class DataMismatchError(Exception):
    """Custom exception for mismatched data lengths."""
    pass

# Data
x = np.arange(0, 11)
y = x ** 2

# Error Handling
if len(x) != len(y):
    raise DataMismatchError("x and y must have the same number of elements.")

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='--', color='b', label='y = x^2')

# Customization
plt.title("Line Plot of y = x^2", fontsize=14, fontweight='bold')
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 11)  # x values from 0 to 10
y = x ** 2  # y values as square of x

# Create Plot
plt.figure(figsize=(8, 5))  # Set figure size
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')  # Line plot

# Customizations
plt.title("Line Plot of y = x²", fontsize=14)  # Title
plt.xlabel("X Values", fontsize=12)  # X-axis label
plt.ylabel("Y Values (Square of X)", fontsize=12)  # Y-axis label
plt.grid(True, linestyle="--", alpha=0.7)  # Add grid
plt.legend()  # Show legend

# Show plot
plt.show()
```

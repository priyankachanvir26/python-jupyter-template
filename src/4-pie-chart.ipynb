### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

class DataMismatchError(Exception):
    """Custom exception for mismatched data lengths."""
    pass

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'lightblue', 'lightcoral', 'lightgreen']

# Error Handling
if len(labels) != len(sizes):
    raise DataMismatchError("Labels and sizes must have the same length.")

# Plot
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True)

# Customization
plt.title("Programming Languages Market Share", fontsize=14, fontweight='bold')
plt.show()
```
